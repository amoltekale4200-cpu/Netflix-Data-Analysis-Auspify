```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Netflix_Cleaned.csv")

print(df.head())
```

```python
df
```

```python
content_type = df['type'].value_counts()

print(content_type)

content_type.plot(kind='bar')
plt.title("Movies vs TV Shows")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.show()
```

```python
rating_count = df['rating'].value_counts().head(10)

print(rating_count)

rating_count.plot(kind='bar')
plt.title("Top 10 Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
year_count = df['release_year'].value_counts().sort_index()

year_count.plot(figsize=(12,5))
plt.title("Netflix Content Release Trend")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
genre = df['listed_in'].str.split(', ').explode()

top_genres = genre.value_counts().head(10)

print(top_genres)

top_genres.plot(kind='bar')
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```
