# Task 4 - Release Year Trend Analysis

## Objective
Analyze Netflix content trends over the years.

## Tools Used
- Python
- Pandas
- Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Netflix_Cleaned.csv")

year_count = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(12,5))
year_count.plot()

plt.title("Netflix Content Release Trend")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")

plt.show()
```

```python
print("Insights:")
print("1. Analyzed content growth over the years.")
print("2. Identified periods with highest content releases.")
print("3. Observed long-term Netflix content trends.")
```
